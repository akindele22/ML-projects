```python
from tweepy import OAuthHandler
from tweepy.streaming import StreamListener
import tweepy
import json
import pandas as pd
import csv
import re
# from Tex
import string
# import prepr
import os
import time
from datetime import datetime
```

```python
j_file = json.load(open('twitter_credentials.json', 'r'))
```

```python
consumer_key = j_file['CONSUMER_KEY']
consumer_secret = j_file['CONSUMER_SECRET']
access_token = j_file['ACCESS_TOKEN']
access_secret = j_file['ACCESS_SECRET']
```

```python
j_file
```

```python
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth)
```

```python
def scrapetweet(search_words, date_since, numTweets, numruns):
    db_tweets = pd.DataFrame(columns=['username', 'acct_desc', 'location', 'following', 'followers', 'total_tweets', 'user_created', 
                                     'tweet_created', 'retweet_count', 'text', 'hashtags'])
    program_start = time.time()
    
    for i in range(0, numruns):
        start_run = time.time()
        
        tweets = tweepy.Cursor(api.search, q=search_words, lang='en', since=date_since, tweet_mode='extended').items(numTweets)
        
        tweet_list = [tweet for tweet in tweets]
        
        num_tweet = 0
        
        for tweet in tweet_list:
            username = tweet.user.screen_name
            acctdesc = tweet.user.description
            location = tweet.user.location
            following = tweet.user.friends_count
            followers = tweet.user.followers_count
            total_tweets = tweet.user.statuses_count
            user_created = tweet.user.created_at
            tweet_created = tweet.created_at
            retweet_count = tweet.retweet_count
            hashtags = tweet.entities['hashtags']
            
            try:
                text = tweet.retweeted_status.full_text
            except AttributeError:
                text = tweet.full_text

            ith_tweet = [username, acctdesc, location, following, followers, total_tweets, user_created,
                         tweet_created, retweet_count,
                        text, hashtags]

            db_tweets.loc[len(db_tweets)] = ith_tweet

            num_tweet += 1

            end_run = time.time()
            duration_run = round((end_run - start_run)/60, 2)

            print(f"no. of tweets scraped for run {i+1} is {num_tweet}.")
            print(f"time taken for {i+1} run to complete is {duration_run}.")

#             time.sleep(30)
        
        
        to_csv_timestamp = datetime.today().strftime('%Y%m%d_%H%M%S')
        path = os.getcwd()
        filename = path + '/data' + to_csv_timestamp + 'dsc_bootcamp_tweets.csv'
        
        db_tweets.to_csv(filename, index=False)
        
    program_end = time.time()

    print("Scraping has completed")
    print("Total time taken to scrap is {} minutes".format(round(program_end - program_start)/60, 2))
```

```python
search_words = "#70daysofML OR #AIbootcamp2020 OR #DSN70daysofML OR #70daysofDL OR #DSN70daysofDL OR #1_million_Al_talents_in_10_year OR #4-days_scorestreak"


date_since = "2020-07-01"
numTweets = 2000
numRuns = 1
```

```python
scrapetweet(search_words, date_since, numTweets, numRuns)
```
